```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("customers.csv")
average_spending = data["Spending"].mean()

customer_spending_greater_than_7000 = data[data["Spending"] > 7000]
customer_from_delhi = data[data["City"] == "Delhi"]

city_wise_average_spending = data.groupby(["City"])["Spending"].mean()
membership_wise_total_spending = data.groupby(["Membership"])["Spending"].sum()
city_wise_customer_count = data.groupby(["City"])["CustomerID"].count()

city_wise_total_spending = data.groupby(["City"])["Spending"].sum()
highest_spending_city = city_wise_total_spending.idxmax()

mst_vlb_membership_type = membership_wise_total_spending.idxmax()

print()
city_wise_total_spending = data.groupby(["City"])["Spending"].sum()
city_wise_total_spending.plot(kind = "bar", figsize = (6, 4))
plt.title("City Wise Total Spending")
plt.xlabel("City")
plt.ylabel("Spending")
plt.show()
print()

membership_wise_total_spending.plot(kind = "bar", figsize = (6, 4))
plt.title("Membership Distribution")
plt.xlabel("Membership")
plt.ylabel("Distribution")
plt.show()
print()

print(f"{highest_spending_city} city has highest spending and {mst_vlb_membership_type} membership customers spend the most")
```

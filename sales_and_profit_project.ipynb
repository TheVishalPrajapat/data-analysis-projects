```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sales_data.csv")

first_5_rows = data.head(5)
total_sales = data["Sales"].sum()
total_profit = data["Profit"].sum()

sales_greater_than_15000 = data[data["Sales"] > 15000]
north_region_data = data[data["Region"] == "North"]

product_wise_total_sale = data.groupby(["Product"])["Sales"].sum()
region_wise_total_profit = data.groupby(["Region"])["Profit"].sum()
product_wise_average_profit = data.groupby(["Product"])["Profit"].mean()

best_selling_product = product_wise_total_sale.idxmax()
most_profitable_region = region_wise_total_profit.idxmax()

print(data)
print()
print(first_5_rows)
print()
print(total_sales)
print()
print(total_profit)
print()
print(sales_greater_than_15000)
print()
print(north_region_data)
print()
print(product_wise_total_sale)
print()
print(region_wise_total_profit)
print()
print(product_wise_average_profit)
print()
print(f"{best_selling_product} is the best selling product and {most_profitable_region} is the most profitable region")

print()
product_wise_total_sale.plot(kind = "bar", figsize=(6,4))
plt.title("Product Wise Total Sale")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
print()

region_wise_total_profit.plot(kind = "bar", figsize=(6,4))
plt.title("Region Wise Total Profit")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

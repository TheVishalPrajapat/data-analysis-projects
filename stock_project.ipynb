```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("stock_data.csv")
first_5_rows = data.head(5)
average_price = data["Price"].mean()

price_greater_than_150 = data[data["Price"] > 150]
stock_abc = data[data["Stock"] == "ABC"]

stock_wise_average_price = data.groupby(["Stock"])["Price"].mean()
stock_wise_total_volume = data.groupby(["Stock"])["Volume"].sum()

highest_stock = stock_wise_average_price.idxmax()
highest_total_volume = stock_wise_total_volume.idxmax()

print("Full Data:-\n",data)
print("\nFirst 5 Rows:-\n",first_5_rows)
print("\nAverage Price:-",average_price)

print("\nPrice Greater Than 150:-\n",price_greater_than_150)
print("\nABC Stock:-\n",stock_abc)

print("\nStock Wise Average Price:-\n",stock_wise_average_price)
print("\nStock Wise Total Volume:-\n",stock_wise_total_volume)

print(f"\n{highest_stock} stock has highest average price and {highest_total_volume} stock has highest trading volume")

print()
stock_wise_average_price.plot(kind = "bar", figsize=(6,4))
plt.title("Stock Wise Average Price")
plt.xlabel("Stock")
plt.ylabel("Price")
plt.show()
print()

stock_wise_total_volume.plot(kind = "bar", figsize=(6,4))
plt.title("Stock Wise Volume")
plt.xlabel("Stock")
plt.ylabel("Volume")
plt.show()
print()

data[data["Stock"] == highest_stock].plot(x="Date", y="Price", kind="line")
plt.title(f"{highest_stock} Analysis")
plt.show()
print()
```
